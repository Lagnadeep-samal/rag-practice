```python
from langchain_community.vectorstores import FAISS
```

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
from langchain_community.retrievers import BM25Retriever
from langchain.retrievers import EnsembleRetriever
from langchain.schema import Document
```

```python
from langchain_community.document_loaders import PyMuPDFLoader
```

```python
loader = PyMuPDFLoader("0-Dataingestparsing/sample.pdf")

docs = loader.load()

print(docs[0].page_content)
```

```python
embedding_model= HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
dense_vectorstore=FAISS.from_documents(docs,embedding_model)
dense_retriever=dense_vectorstore.as_retriever()
```

```python
sparse_retriever=BM25Retriever.from_documents(docs)
sparse_retriever.k=3
hybrid_retriever=EnsembleRetriever(retrievers=[dense_retriever,sparse_retriever],
                                   weights=[0.8,0.2]
                                   )
```

```python
query = "What is Error Control Coding and why is it used in communication systems?"

results = hybrid_retriever.invoke(query)

# Step 10: Print results
for i, doc in enumerate(results):
    print(f"\n🔹 Document {i+1}:\n{doc.page_content}")
```
